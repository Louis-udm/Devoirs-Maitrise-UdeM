We adjust the value of the Critic($max=15$) dynamically during the training. For each minibatch, when fixed the generator and training the discriminator, we want the discriminator's loss, $mean(real\_data\_loss+fake\_data\_loss)<0.69$, that is $E[p]>0.5$ ; similarly when fixed the discriminator, we want the generator to generate The probability of the picture being judged by discriminator, E[p]>0.5. This can make discriminator and generator work together as much as possible. Experiments have found that the effect is good.

### comments:
1. The dcgan algorithm is very unstable during training. It is often the case that a gradiant descent is too fast. For example, the generator quickly becomes too powerful, causing the discriminator to fail to train, or vice versa. We tried to improve this situation by dynamically controlling the critic parameter.
2. If you adjust the parameters and monitor the d and g loss, you can generate a good picture. For example, we dynamically control the tritic cycle parameter based on the loss, which can greatly increase the stability.
2. Nearby is more unstable than deconvolution, and bilinear is the hardest to train. It is slow.
3. Bilinear and nearest are easy to collapsed, while bilinear is the easiest to cause collapse.
After every 1-2epoch, bilinear will completely collapse to another face. It is slowest.
4. The "nearest" is also similar, but bilinear is more obvious.
5. The wgan algorithm is much more stable than the dc algorithm when training.
4. Wgan's advantages: fast and stable training. Disadvantages: The resulting picture is not better than DCGAN, and the resulting face often has facial defective features.
